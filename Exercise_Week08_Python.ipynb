### Question 03

Create a random 7×24 matrix representing hourly temperatures for one week.

(7 days, 24 Hour)

* Find the hour with the highest average temperature.
* Find the day with the highest average temperature.
* Compute the difference between max and min temperature.

```python
array_03 =np.random.randint(35,40,(7,24)) 
AverHourTemp=array_03.mean(axis=0)
HighestHourTemp = AverHourTemp.max()
Hour = np.argmax(AverHourTemp + 1)
print(f"THE HOGHEST AVERAGE TEMPTURE IS : {HighestHourTemp} AND IT IS AT {Hour} O'CLOCK")
AverDayTemp = array_03.mean(axis=1)
HighestDayTemp = AverDayTemp.max()

Day = np.argmax(AverDayTemp)

def WeekDayCalc (day):
    if day == 0 :
        return "Saturday"
    elif day == 1:
        return "Sunday"
    elif day == 2:
        return "Monday"
    elif day == 3:
        return "Tuesday"
    elif day == 4:
        return "Wednesday"
    elif day == 5:
        return "Thursday"
    elif day == 6:
        return "Friday"
    else:
        print("invalid day")

print(f"THE HOGHEST AVERAGE TEMPTURE IS : {HighestDayTemp} AND IT IS IN {WeekDayCalc(Day)}")

MaxTemp = array_03.max()
MinTemp = array_03.min()
DifMinaMax = MaxTemp - MinTemp
print(f"DIFFERENT OF MINIMMUM TEMPTURE AN MAXIMMUM IS : {DifMinaMax}")

```

### Question 04
Create a 30×5 matrix (30 students, 5 courses).

* Compute each student's average.
* Compute each course average.
* Identify students who scored above 15 in at least 3 courses.
* Find the hardest course (with the loest scores)

```python
array_04 = np.random.randint(10,20,(30,5))
StudentAver = array_04.mean(axis=1)
print(StudentAver)

CourseAver = pd.Series((array_04.mean(axis=0)).tolist())
CourseAver.index = ["First Course","Second Course","Third Course","Forth Course","Fifth Course"]
print(CourseAver)

AcceptedScore = np.where(array_04>15,1,0)
counter = AcceptedScore.sum(axis=1)
AcceptedCourse = np.where(counter >= 3)
NameList=[]
def StudentName (array,list):
    for n in range(len(array)):
        list.append(f"student {n+1}")
StudentName(array_04,NameList)
StudentsTable = pd.Series(array_04.tolist(),name="Students")
StudentsTable.index = NameList
AcceptedStudents = StudentsTable.index[AcceptedCourse]
print(f"The accepted students are {StudentsTable.index[AcceptedCourse].tolist()}")

HradestCourse = np.argmin(CourseAver)
print(f"The hardest score is {CourseAver.index[HradestCourse]} by {CourseAver.min()} score")

```

Income array (values between 5000–15000)

Expenses array (values between 3000–12000)

* Identify loss months (profit < 0).

### Question 08

Create a Series of 50 people birthday year (1925 to 2025)

and set indexes with their gender (random Female/Male).

then, separate these 50 pople into 6 different series based on their generation:

* Alpha: Born 2013–2025.
* Z: Born 1997–2012 (Ages 14–29).
* Millennials: Born 1981–1996 (Ages 30–45).
* X: Born 1965–1980 (Ages 46–61).
* Baby Boomers: Born 1946–1964 (Ages 62–80).
* Silent: Born 1925–1945 (Ages 81–101). 

Finally, print number of Female and Male popele in each generation.

```python
PeopleSerie = pd.Series(np.random.randint(1945,2025,50))
gender = np.random.choice(["Female","Male"],50)
PeopleSerie.index = gender
generations = ["Alpha","Z","Millennials","X","Baby Boomers","Silent"]
Alpha = pd.Series(PeopleSerie.iloc[np.where((PeopleSerie>=2013)&(PeopleSerie<=2025))], name="Alpha")
Z = pd.Series(PeopleSerie.iloc[np.where((PeopleSerie>=1997)&(PeopleSerie<=2012))], name="Z")
Millennials = pd.Series(PeopleSerie.iloc[np.where((PeopleSerie>=1981)&(PeopleSerie<=1996))], name="Millennials")
X = pd.Series(PeopleSerie.iloc[np.where((PeopleSerie>=1965)&(PeopleSerie<=1980))], name="X")
BabyBoomers = pd.Series(PeopleSerie.iloc[np.where((PeopleSerie>=1946)&(PeopleSerie<=1964))], name="Baby Boomers")
Silent = pd.Series(PeopleSerie.iloc[np.where((PeopleSerie>=1925)&(PeopleSerie<=1945))], name="Silent")

FemalecounterList = []
def GenCounter (serie,text,list):
    counterGen = np.where(serie.index == text , 1 , 0)
    list.append(counterGen.sum().tolist())
GenCounter(Alpha,"Female",FemalecounterList)
GenCounter(Z,"Female",FemalecounterList)
GenCounter(Millennials,"Female",FemalecounterList)
GenCounter(X,"Female",FemalecounterList)
GenCounter(BabyBoomers,"Female",FemalecounterList)
GenCounter(Silent,"Female",FemalecounterList)
FemaleGenSerie = pd.Series(np.array(FemalecounterList),name= "Female")
FemaleGenSerie.index = generations
print(FemaleGenSerie)
print(f"*" * 30)

malecounterList = []
GenCounter(Alpha,"Male",malecounterList)
GenCounter(Z,"Male",malecounterList)
GenCounter(Millennials,"Male",malecounterList)
GenCounter(X,"Male",malecounterList)
GenCounter(BabyBoomers,"Male",malecounterList)
GenCounter(Silent,"Male",malecounterList)
maleGenSerie = pd.Series(np.array(malecounterList),name= "Male")
maleGenSerie.index = generations
print(maleGenSerie)





```
