### Question 01
Find the top 5 countries with the highest population density (Density P/Km2) but with a land area (Land Area Km2) greater than 100,000 square kilometers.

```python
density = df_population["Population"] / df_population["Land Area (Km²)"]
df_population.fillna({"Density (P/Km²)": density },inplace=True)
df_population
```

```python

Q1DfPopulation = df_population.sort_values('Density (P/Km²)',ascending=False).loc[df_population['Land Area (Km²)']>100000,:].head(5)
Q1DfPopulation
```

### Question 04
Find the top 5 countries with the highest natural population growth. 

* Hint: Natural Growth equals Net Change minus Net Migrants.

```python
NaturalGrorth = (df_population["Net Change"] - df_population["Migrants (net)"]).sort_values(ascending=False).head(5)
df_population.iloc[NaturalGrorth.index.tolist()]
```

### Question 05

Find the country with the largest land area that has a negative net migration (net migration: means sends out more migrants than it receives)

```python
NegativeMigrationC = df_population.loc[[(df_population["Land Area (Km²)"].idxmax()) and (df_population["Migrants (net)"].idxmin())],:]
NegativeMigrationC
```

### Question 07

Identify the top 20 countries with the highest World Share
population percentage.
* Which continent are these contries in?

```python
worldShare = ((df_population['Population'] / df_population['Population'].sum()) * 100).round(3)
df_population.fillna({"World Share": worldShare},inplace=True)
```

```python
HighestWorldShare = df_population.sort_values("World Share",ascending=False).head(20)
HighestWorldShare
```

### Question 08

Plot a histogram with 20 bins for the median age of the countries, and draw a dashed red vertical line to indicate the global mean.

```python
plt.hist(df_population['Median Age'],rwidth=0.6,bins=20)
plt.axvline(x = float(df_population['Median Age'].mean()),linestyle = "--",color="r")
plt.legend(["Wolrd Mean","Country Median Age"])
```

### Question 09

Create a plot with fertility rate on the X-axis and median age on the Y-axis.

```python
plt.scatter(df_population['Fert. Rate'],df_population['Median Age'],color="g")
plt.xlabel("Fertility Rate")
plt.ylabel("Median Age")
```

### Question 11

Create a line plot showing the population of the top 10 largest countries by land area. The X-axis should be sorted by land area in descending order, and the line should have circular markers.


```python
LargeCountry=df_population.sort_values("Land Area (Km²)",ascending=False).head(10)
LargeCountry
```

```python

plt.plot(LargeCountry['Land Area (Km²)']/10000,LargeCountry['Population']/1000000,marker="o",mfc="r",mec="r",linestyle="-",color="m")
plt.xlabel("Land Area (Km²)(10000)")
plt.ylabel("Population(Millions)")
```
